```python
import pandas as pd
```

```python
df = pd.read_csv('placement.csv')
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize = (5,4 ))
plt.scatter(x = 'cgpa',y = 'package', data = df)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
input_data = df[['cgpa']]
output_data = df['package']
```

```python
x_train,x_test,y_train,y_test = train_test_split(input_data,output_data, test_size = 0.2, random_state = 52)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_train,y_train)*100
```

```python
lr.predict([[6.89]]).round()
```

```python
(lr.coef_*6.89)+(lr.intercept_)
```
