```python
import pandas as pd 
```

```python
df = pd.read_csv('loan.csv')
df.head(3)
```

```python
df.describe()
```

```python
q1 = df['CoapplicantIncome'].quantile(0.25)
q3 = df['CoapplicantIncome'].quantile(0.75)
iqr = q3 - q1
q1,q3,iqr
```

```python
lower_bound = q1 - (1.5*iqr)
upper_bound = q3 + (1.5*iqr)
lower_bound,upper_bound
```

```python
iqr_df = df[(df['CoapplicantIncome']<=upper_bound) & (df['CoapplicantIncome']>=lower_bound)]
```

```python
iqr_df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize = (10,4))
plt.subplot(1,2,1)
plt.title('Before Removing outliers')
sns.distplot(df['CoapplicantIncome'])

plt.subplot(1,2,2)
plt.title('After Removing outliers')
sns.distplot(iqr_df['CoapplicantIncome'])
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
nml = MinMaxScaler()
nml.fit(iqr_df[['CoapplicantIncome']])
```

```python
iqr_df['CoapplicantIncome_nml'] = nml.transform(iqr_df[['CoapplicantIncome']])
```

```python
plt.figure(figsize = (10,4))
plt.subplot(1,2,1)
plt.title('Before')
sns.distplot(iqr_df['CoapplicantIncome'])

plt.subplot(1,2,2)
plt.title('After')
sns.distplot(iqr_df['CoapplicantIncome_nml'])
```
